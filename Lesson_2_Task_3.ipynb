```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
ax = plt.gca()
t = np.linspace(0, 2 * np.pi, 100)
r = 1
x = r * np.cos(t)
y = r * np.sin(t)
ax.plot(x, y)
plt.axis('equal')
plt.show()
```

```python
ax = plt.gca()
l = np.linspace(0, 2 * np.pi, 100)
r = 1
x = r * np.cos(l)
y = r * np.sin(l)
ax.plot(x, y)
plt.show()
```

```python
x = np.linspace(1, 20, 100)
y = 1/x
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
