```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 5*Y
X1 = X*2
Y1 = Y*2
X1 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid(X1, Y1)
Z1 = 2*(X1+20) + 5*(Y1+20)
ax.plot_surface(X, Y, Z,)
ax.plot_surface(X1, Y1, Z1, color='g', alpha=0.5)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from math import sin, cos, pi
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
t = np.linspace(-2*pi,2*pi, 100)
Z = np.linspace(0,1,100)
Z,t = np.meshgrid(Z, t)
R = 0.5
X = (R*np.cos(t))
Y = (R*np.sin(t))
ax.plot_surface(X,Y,Z,)
plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X2 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid(X1, Y1)
X2, Y2 = np.meshgrid(X2, Y2)
Z1 = (8 - 2*X1 - 3*Y1)**4
Z2 = (2*X2+1 + 3*Y2+2)**3
ax.plot_surface(X1, Y1, Z1)
ax.plot_surface(X2, Y2, Z2)
plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection = '3d')
u = np.linspace(0, 2 * np.pi*-12/13, 100)
v = np.linspace(0, np.pi, 100)
x = 5 * np.outer(np.cos(u), np.sin(v))
y = 5 * np.outer(np.sin(u), np.sin(v))
z = 5 * np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, color='g')
ax.view_init(30,0)
plt.show()
```
