```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 201)
a1, a2, a3 = 2, 3, 4
k1, k2, k3 = -2, 4, -1
b1, b2, b3 = 3, -5, 2
plt.plot(x, np.cos(x-a1)*k1+b1)
plt.plot(x, np.cos(x-a2)*k2+b2)
plt.plot(x, np.cos(x-a3)*k3+b3)
plt.xlabel('x')
plt.xlabel('y')
plt.grid(True)
plt.show()
```

```python
print ('Для отображения графиков функции: y = cos(x-a), y = k*cos(x-a), y = k*cos(x-a)+b')
a = input ('1.Введите числовой параметр (a) функции: ')
b = input ('2.Введите числовой параметр (b) функции: ')
k = input ('3.Введите числовой параметр (k) функции: ')
x = np.linspace(-np.pi, np.pi, 201)
a = int(a)
k = int(k)
b = int(b)
plt.plot(x, np.cos(x-a))
plt.plot(x, np.cos(x-a)*k)
plt.plot(x, np.cos(x-a)*k+b)
plt.xlabel('x')
plt.xlabel('y')
plt.grid(True)
plt.show()
```
