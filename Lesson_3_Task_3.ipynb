```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#перевести полярные координаты в декартовы.
phi = np.arange(1, 5, 10)
rho = 2*phi
x = rho *np.cos(phi)
y = rho *np.sin(phi)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
print(x)
print(y)
plt.scatter(x, y, color='r', s=90, marker='o') #точка координат
plt.show()
```

```python
#график отрезка прямой линии в полярных координатах.
fig1=plt.figure(figsize=(5,5))
ax = plt.subplot(111,projection='polar')
phi = np.pi*2
rho = [0, 2*phi/3]
ax.plot(rho,color='r', lw=5)
ax.grid(True)
print(rho)
plt.show()
```
