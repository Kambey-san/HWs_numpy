### 2. Найдите псевдорешение:


```python
import numpy as np
from numpy import linalg as la
```

```python
a = np.array([[1, 2, -1], [3, -4, 0], [8, -5, 2], [2, 0, -5], [11, 4, -7]])
b = np.array([1, 7, 12, 7, 15])
x = np.linalg.lstsq(a, b, rcond=None)
print ('Псевдорешение системы:', x)
```
